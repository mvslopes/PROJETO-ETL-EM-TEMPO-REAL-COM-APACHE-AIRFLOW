##**Construindo o processo de ETL - Extração, transformação e carga de dados em Tempo Real com Apache Airflow**


pipeline_encanamento.png



Neste projeto, o objetivo foi desenvolver um processo de **Extração, Transformação e Carregamento de dados em um Data Warehouse** para a empresa fictícia de varejo denominada **LogSolutions**. A empresa buscava entender os dados específicos da área de **logística** para melhorar a tomada de decisões.

**As tecnologias usadas no projeto foram:**

* Apache Airflow;
* linguagem de programação Python;
* Anaconda Python;
* PostgreSQL;
* Docker.

O **Apache Airflow** foi utilizado para agendar um horário específico do dia e monitorar uma pasta na rede da empresa, onde os arquivos CSV gerados pelos sistemas transacionais foram armazenados.

Sempre que novos arquivos eram adicionados a essa pasta, o processo ETL lia, limpava, processava, organizava e carregava esses arquivos no Data Warehouse.

##**Documentando o Projeto**


**1. Breve Introdução**

A **LOGSOLUTIONS** estava empenhada em aproveitar oportunidades de aprendizado e aprimoramento ao implementar um **Data Warehouse (DW)** direcionado para a área de **logística**. O principal objetivo era reduzir os custos de frete e aprimorar a eficiência das operações de entrega. Este documento apresenta uma visão do projeto, as soluções propostas e os resultados esperados.

**1.2 Detalhamento do Problema e Impacto no Negócio**

**Desafios:**

* Altos custos de frete;
* Necessidade de avaliação do desempenho das transportadoras;
* Diferença entre a data de coleta e a data de entrega como um fator crítico;
* Impacto direto nos custos operacionais.


**Objetivos:**

* Reduzir os custos de frete;
* Avaliar e melhorar o desempenho das transportadoras;
* Minimizar a diferença entre a coleta e a entrega;
* Aprimorar a eficiência geral das operações logísticas;
* Aumentar a rentabilidade da empresa.


**1.3 Métricas de Resultados**

* Redução dos custos de frete;
* Aprimoramento da eficiência das entregas;
* Análise comparativa do desempenho das transportadoras;
* Mensuração do intervalo de tempo entre a coleta e a entrega dos produtos.

###**2- SOLUÇÃO PROPOSTA**

**2.1 Tecnologias Utilizadas**

A solução proposta para os desafios de negócios da LOGSOLUTIONS envolveu:

* A implementação de um **Data Warehouse (DW)** baseado no **PostgreSQL**, que era uma plataforma robusta e escalável.

* Além disso, a **automação das tarefas de Extração, Transformação e Carregamento (ETL)** foi realizada por meio do **Apache Airflow**, um framework confiável para agendar e executar processos ETL de forma eficaz;

* Desenvolvimento em **linguagem Python dos códigos para ETL** versáteis e de fácil manutenção.

* A infraestrutura foi suportada pelo **Docker**.


**2.2 Desafios Técnicos**

Os desafios técnicos deste projeto englobaram:

**Modelagem de Data Mart:** O primeiro desafio foi projetar adequadamente o Data Mart, definindo a estrutura e os relacionamentos dos dados para atender às necessidades da área de logística.

**Obs:** Um data mart é uma forma simples de data warehouse que se concentra em um único assunto ou linha de negócios, como vendas, finanças ou marketing.No nosso projeto a nossa linha de negócio foi a Logística.

**Integração de Dados:** Foi fundamental integrar dados de várias fontes, como bancos de dados transacionais e outras fontes de informações, garantindo a consistência e qualidade dos dados.

**Desenvolvimento de ETL Eficientes:** Criar processos ETL eficientes foi crucial, envolvendo a coleta, transformação e carregamento de dados de forma ágil e precisa.

###**3- RESULTADO**

**3.1 Insights e Conhecimento Gerado**

A expectativa é que o projeto forneça insights significativos sobre os processos logísticos da empresa, identificando oportunidades de aprimoramento e implementando medidas eficazes de redução de custos.
