##**SPRINT 04 - Ativação da DAG e alimentação do Data Warehouse**

O principal objetivo da **Sprint 04** era a ativação da DAG, referindo-se ao processo de iniciar a execução ou agendamento de uma DAG (Directed Acyclic Graph ou Grafo Direcionado Acíclico).

A ativação da DAG envolveu o seguinte processo:

**Início da Execução:**

**Monitoramento e Registro**: Durante a execução da **DAG, o Airflow** monitorou o progresso de cada tarefa. Os resultados e os registros foram armazenados para referência futura.

**Notificação de Resultados**: Após a conclusão da execução da DAG, recebemos notificações sobre o status da execução, como sucesso, falha ou outras informações relevantes.

**Manutenção e Aprimoramento** : Com base nos resultados e no monitoramento, foram feitas ajustes na DAG, como correções de erros, otimizações de desempenho ou aprimoramentos de funcionalidades.

Em resumo, a ativação da DAG foi o processo de colocar em funcionamento o fluxo de trabalho definido na DAG, garantindo que as tarefas fossem executadas de acordo com a lógica e as dependências especificadas. Isso foi fundamental para automatizar processos, agendar tarefas e garantir a consistência e confiabilidade das operações em sistemas de gerenciamento de fluxos de trabalho.

**2.1**- Por que isso aconteceu?

---

A Reposta está na mensagem de erro. o **caminho é / opt/airflow/dags/my_file.py.** esse que devenos colocar no código **Python para ETL**. O diretório que ele está procurando na prática é um diretório interno no contêiner Docker.

Esse problema só acontece devido ao uso o **Docker**. Com Apache Airflow instalado localmente esse problema não ocorre e podemos usar o caminho direto que foi usado nas tentativas anteriotes.

No entanto, quando o Docker foi usado para citar o contêiner do Airflow, ele fez o mapeamento de volumes. Portanto, essa pasta de DAGs foi criada na máquina local está mapeada para / opt/airflow/dags no container do Airflow.

**Para verificar e garantir isso, acesse o contêiner do webserver no Docker.**



locaal_Dags4.png

---

Perceba que é o mesmo conteúdo da pasta local. O que aconteceu foi que, quando o **Docker** criou o contêiner e instalou o **Apache Airflow**, fez o mapeamento. Portanto, ele viu os arquivos, mas, na hora de executar o **script Python**, não olhou para a máquina local. Ele olhou para o mapeamento que estava no contêiner Docker.

**Portanto, para resolver esse problema**, foi editados todos os scripts Python colocando esse novo endereço. Perceba que é o mesmo **conteúdo da pasta local**. O que aconteceu é que, quando o Docker criou o contêiner e instalou o **Apache Airflow**, fez o mapeamento. Portanto, ele viu os arquivos, mas, na hora de executar o script Python, não olhou para a máquina local. Ele olhou para o mapeamento que estava no contêiner Docker.

**Portanto, para resolver esse problema, você precisou copiar essa pasta de DAGs que estava no contêiner Docker. Em seguida, editou todos os scripts Python com esse endereço. / opt/airflow/dags/dados**
